```python
import numpy as np
from tenpy.networks.mps import MPS
from tenpy.models.hubbard import FermiHubbardModel
from tenpy.algorithms import dmrg

import pickle
import matplotlib.pyplot as plt
```

```python
def run_dmrg_half_filling(L, U, t, mu, chi_max, sweeps, hc_flag=False):
    model_params = {
        'L': L,
        'U': U,
        't':t,
        'mu': mu,
        'bc_MPS': 'finite',
        'bc_x':'periodic',
        'explicit_plus_hc':hc_flag
    }

    model = FermiHubbardModel(model_params)

    # Initialize the MPS for half-filling with an alternating pattern of up and down spins
    initial_state = ['up', 'down'] * (L // 2) if L % 2 == 0 else ['up', 'down'] * (L // 2) + ['up']
    psi = MPS.from_product_state(model.lat.mps_sites(), initial_state, bc='finite')

    dmrg_params = {
        'mixer': True,  # Enables mixing for better convergence in difficult phases
        'trunc_params': {
            'chi_max': chi_max,
            'svd_min': 1.e-10
        },
        'max_sweeps': sweeps,
    }

    eng = dmrg.TwoSiteDMRGEngine(psi, model, dmrg_params)
    E0, psi = eng.run()

    return E0
```

```python
from scipy.integrate import quad
from scipy.special import j0, j1
import numpy as np

def integrand(omega, U):
    return j0(omega) * j1(omega) / (omega * (1 + np.exp(2 * U * omega)))

# Define the integral as a function of U
def bethe_ground_state(U):
    result, _ = quad(lambda x: integrand(x, U), 0, np.inf)
    return result*4*(-1)

# Test the function for a specific value of U
bethe_ground_state(U/4)
```

```python
# Parameters
L = 8  # Number of lattice sites (ensure it is even for a simple setup)
U = 4.0  # On-site interaction strength
t = np.exp(1j*np.pi*0.1)  # Hopping term
mu = 0  # Chemical potential (often set to 0 for simplicity at half filling)
chi_max = 200  # Maximum bond dimension
sweeps = 100  # Number of DMRG sweeps



ground_state_energy = run_dmrg_half_filling(L, U, t, mu, chi_max, sweeps)
print("Ground State Energy at Half Filling:", ground_state_energy/L)
```

```python
phase_vals = np.arange(0, np.pi/4, 0.1)
hc_true_vals = np.zeros_like(phase_vals)
hc_false_vals = np.zeros_like(phase_vals)

for i, phase in enumerate(phase_vals):
    t = np.exp(1j*phase)
    hc_false_vals[i] = run_dmrg_half_filling(L, U, t, mu, chi_max, sweeps, False)
    hc_true_vals[i] = run_dmrg_half_filling(L, U, t, mu, chi_max, sweeps, True)
```

```python
k_num = 10
U = 0.0  # On-site interaction strength
t = 0.0  # Hopping term
mu = 0  # Chemical potential (often set to 0 for simplicity at half filling)
chi_max = 1000  # Maximum bond dimension
sweeps = 1000  # Number of DMRG sweeps



norb_list = np.arange(4, 7, 2)

twist_avg_energy = np.zeros(len(norb_list))
std_energy = np.zeros(len(norb_list))
min_energy = np.zeros(len(norb_list))

for ind, norb in enumerate(norb_list):
    energy_arr = np.zeros(k_num)
    for kx in range(-int(k_num/2), int(k_num/2)):
        t_star = np.exp(1j*kx/k_num*np.pi)
        ground_state_energy = run_dmrg_half_filling(norb, U, t_star, mu, chi_max, sweeps)/norb
        energy_arr[kx+int(k_num/2)] = ground_state_energy

    twist_avg_energy[ind] = energy_arr.mean()
    min_energy[ind] = energy_arr.min()
    std_energy[ind] = run_dmrg_half_filling(norb, U, 1, mu, chi_max, sweeps)/norb

    print(norb)
```

```python
k_num = 10
mu = 0  # Chemical potential (often set to 0 for simplicity at half filling)
chi_max = 500  # Maximum bond dimension
sweeps = 500  # Number of DMRG sweeps

U_range = np.arange(2, 9, 1)
norb_list = np.arange(4, 22, 2)

twist_avg_energy = np.zeros(shape=(len(U_range), len(norb_list)))
std_energy = np.zeros(shape=(len(U_range), len(norb_list)))

for U_ind, U in enumerate(U_range):
    for ind, norb in enumerate(norb_list):
        energy_arr = np.zeros(k_num)
        for kx in range(-int(k_num/2), int(k_num/2)):
            t_star = np.exp(1j*kx/k_num*np.pi)
            ground_state_energy = run_dmrg_half_filling(norb, U, t_star, mu, chi_max, sweeps)/norb
            energy_arr[kx+int(k_num/2)] = ground_state_energy

        twist_avg_energy[U_ind, ind] = energy_arr.mean()
        std_energy[U_ind, ind] = run_dmrg_half_filling(norb, U, 1, mu, chi_max, sweeps)/norb
        print(U, norb)
```

```python
data_dict = {}
data_dict['U_range'] = U_range
data_dict['norb_list'] = norb_list
data_dict['twist_avg'] = twist_avg_energy
data_dict['std_DMRG'] = std_energy

with open('DMRG_OHK_set2_k10.pkl', 'wb') as handle:
    pickle.dump(data_dict, handle, protocol=pickle.HIGHEST_PROTOCOL)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Assuming U_range, std_energy, twist_avg_energy, norb_list, and bethe_ground_state are defined

bethe_list = np.zeros(len(U_range))
for U_ind, U in enumerate(U_range):
    bethe_list[U_ind] = bethe_ground_state(U / 4)

# Create a subplot grid with 1 row and 2 columns
fig, axes = plt.subplots(1, 2, figsize=(10, 5), sharey=True)

axes[0].set_ylabel("Relative error from Bethe Ansatz")
# Plot for std_energy
for ind, norb in enumerate(norb_list):
    axes[1].plot(U_range, np.abs((std_energy[:, ind] - bethe_list) / bethe_list), marker='o', label=f'L = {norb}')
axes[1].set_xlabel("U")
axes[1].set_ylabel("Relative error from Bethe GSE")
axes[1].set_title("Standard DMRG")
axes[1].legend()
axes[1].set_ylim(0, 0.08)

# Plot for twist_avg_energy
for ind, norb in enumerate(norb_list):
    axes[0].plot(U_range, np.abs((twist_avg_energy[:, ind] - bethe_list) / bethe_list), marker='^', label=f'norb = {norb}')
axes[0].set_xlabel("U")
axes[0].set_title("Twist Average")
axes[0].legend()
axes[0].set_ylim(0, 0.08)
plt.tight_layout()
plt.show()

```
